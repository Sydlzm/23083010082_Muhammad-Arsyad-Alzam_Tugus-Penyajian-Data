### Deskripsi Data
#### Dataset ini berisi data tentang pendidikan global dan beberapa indikator yang relevan untuk mengukur kualitas atau keadaan pendidikan di sebuah negara. Variabel-variabel yang diamati adalah, yakni sebagai berikut:
#### 1. Countries and Areas: Nama negara dan wilayah.
#### 2. Latitude: Koordinat garis lintang dari lokasi geografis.
#### 3. Longitude: Koordinat garis bujur dari lokasi geografis.
#### 4. OOSR_Pre0Primary_Age_Male: Tingkat anak laki-laki usia prasekolah yang tidak bersekolah.
#### 5. OOSR_Pre0Primary_Age_Female: Tingkat anak perempuan usia prasekolah yang tidak bersekolah.
#### 6. OOSR_Primary_Age_Male: Tingkat anak laki-laki usia sekolah dasar yang tidak bersekolah.
#### 7. OOSR_Primary_Age_Female: Tingkat anak perempuan usia sekolah dasar yang tidak bersekolah.
#### 8. OOSR_Lower_Secondary_Age_Male: Tingkat anak laki-laki usia sekolah menengah pertama yang tidak bersekolah.
#### 9. OOSR_Lower_Secondary_Age_Female: Tingkat anak perempuan usia sekolah menengah pertama yang tidak bersekolah.
#### 10. OOSR_Upper_Secondary_Age_Male: Tingkat anak laki-laki usia sekolah menengah atas yang tidak bersekolah.
#### 11. OOSR_Upper_Secondary_Age_Female: Tingkat anak perempuan usia sekolah menengah atas yang tidak bersekolah.
#### 12. Completion_Rate_Primary_Male: Tingkat penyelesaian pendidikan sekolah dasar oleh anak laki-laki.
#### 13. Completion_Rate_Primary_Female: Tingkat penyelesaian pendidikan sekolah dasar oleh anak perempuan.
#### 14. Completion_Rate_Lower_Secondary_Male: Tingkat penyelesaian pendidikan sekolah menengah pertama oleh anak laki-laki.
#### 15. Completion_Rate_Lower_Secondary_Female: Tingkat penyelesaian pendidikan sekolah menengah pertama oleh anak perempuan.
#### 16. Completion_Rate_Upper_Secondary_Male: Tingkat penyelesaian pendidikan sekolah menengah atas oleh anak laki-laki.
#### 17. Completion_Rate_Upper_Secondary_Female: Tingkat penyelesaian pendidikan sekolah menengah atas oleh anak perempuan.
#### 18. Grade_2_3_Proficiency_Reading: Kemampuan membaca siswa kelas 2-3.
#### 19. Grade_2_3_Proficiency_Math: Kemampuan matematika siswa kelas 2-3.
#### 20. Primary_End_Proficiency_Reading: Kemampuan membaca pada akhir pendidikan dasar.
#### 21. Primary_End_Proficiency_Math: Kemampuan matematika pada akhir pendidikan dasar.
#### 22. Lower_Secondary_End_Proficiency_Reading: Kemampuan membaca pada akhir pendidikan menengah pertama.
#### 23. Lower_Secondary_End_Proficiency_Math: Kemampuan matematika pada akhir pendidikan menengah pertama.
#### 24. Youth_15_24_Literacy_Rate_Male: Tingkat literasi pemuda laki-laki usia 15-24 tahun.
#### 25. Youth_15_24_Literacy_Rate_Female: Tingkat literasi pemuda perempuan usia 15-24 tahun.
#### 26. Birth_Rate: Tingkat kelahiran di negara atau wilayah terkait.
#### 27. Gross_Primary_Education_Enrollment: Tingkat partisipasi bruto dalam pendidikan dasar.
#### 28. Gross_Tertiary_Education_Enrollment: Tingkat partisipasi bruto dalam pendidikan tinggi.
#### 29. Unemployment_Rate: Tingkat pengangguran di negara atau wilayah terkait.

```python
selected_features = [
    'Completion_Rate_Primary_Male', 'Completion_Rate_Primary_Female',
    'Youth_15_24_Literacy_Rate_Male', 'Youth_15_24_Literacy_Rate_Female',
    'Birth_Rate', 'Gross_Primary_Education_Enrollment'
]

n_cols = 2
n_rows = (len(selected_features) + n_cols - 1) // n_cols  

plt.figure(figsize=(15, 3 * n_rows))

for i, feature in enumerate(selected_features, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.scatterplot(x=feature, y='Unemployment_Rate', data=df)
    plt.title(f'Scatter Plot: {feature} vs Unemployment Rate')
    plt.xlabel(feature)
    plt.ylabel('Unemployment Rate')

plt.tight_layout()

plt.show()
```

```python
variables = numeric_columns

n_cols = 3
n_rows = (len(variables) + n_cols - 1) // n_cols

plt.figure(figsize=(15, 3* n_rows))  

for i, var in enumerate(variables, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.histplot(df[var], kde=True)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
variables = numeric_columns

top_10_lists = {}

for var in variables:
    top_10 = df.groupby('Countries and areas')[var].sum().reset_index().sort_values(by=var, ascending=False).head(10)
    top_10_lists[var] = top_10

n_cols = 2 
n_rows = (len(variables) + n_cols - 1) // n_cols 

plt.figure(figsize=(10, 6 * n_rows))

for i, var in enumerate(variables):
    plt.subplot(n_rows, n_cols, i+1)
    top_10 = top_10_lists[var]
    plt.bar(top_10['Countries and areas'], top_10[var])
    plt.xlabel('Countries')
    plt.ylabel(f'{var}')
    plt.title(f'Top 10 {var}')
    plt.xticks(rotation=45, ha='right') 

plt.tight_layout()  
plt.show()
```
